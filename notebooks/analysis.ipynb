# Yelp Academic Dataset - Exploratory Data Analysis (EDA)

This notebook performs an Exploratory Data Analysis on the Yelp Academic Dataset (Business, Checkins, and Tips). This analysis serves as the foundation for the interactive dashboard and uncovers high-level patterns in ratings, user engagements, and geographical coverages.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set matplotlib styling
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
```

## 1. Loading the Cleaned Dataset

We load the consolidated business records and aggregated time-series datasets that were preprocessed from the raw Yelp JSON files.

```python
data_dir = r"../data"
biz_path = os.path.join(data_dir, "cleaned_businesses.csv")
ts_path = os.path.join(data_dir, "cleaned_time_series.csv")

df_biz = pd.read_csv(biz_path)
df_ts = pd.read_csv(ts_path)

print(f"Businesses: {df_biz.shape[0]} rows, {df_biz.shape[1]} columns")
print(f"Time Series: {df_ts.shape[0]} rows, {df_ts.shape[1]} columns")
```

## 3. Rating & Review Analysis

Let's look at the distribution of stars ratings and how review counts differ.

```python
df_biz[["stars", "review_count", "checkin_count", "tip_count"]].describe()
```

```python
# Plot Stars Rating Distribution
plt.figure(figsize=(7, 4))
sns.histplot(df_biz['stars'], bins=np.arange(0.75, 5.75, 0.5), color='skyblue', edgecolor='black')
plt.title('Distribution of Star Ratings')
plt.xlabel('Stars')
plt.ylabel('Count')
plt.xticks(np.arange(1.0, 5.5, 0.5))
plt.show()
```

## 4. Top Cities & Categories

Identifying where the businesses are concentrated and what their major industries are.

```python
# Top 10 cities by business counts
plt.figure(figsize=(8, 4))
sns.countplot(data=df_biz, y='city', order=df_biz['city'].value_counts().head(10).index, palette='viridis')
plt.title('Top 10 Cities by Number of Businesses')
plt.xlabel('Business Count')
plt.show()
```

```python
# Business counts by state
plt.figure(figsize=(8, 4))
sns.countplot(data=df_biz, x='state', order=df_biz['state'].value_counts().head(10).index, palette='magma')
plt.title('Top 10 States by Number of Businesses')
plt.ylabel('Business Count')
plt.show()
```

```python
# Distribution of Primary Categories
plt.figure(figsize=(8, 5))
sns.countplot(data=df_biz, y='primary_category', order=df_biz['primary_category'].value_counts().index, palette='plasma')
plt.title('Distribution of Primary Industry Categories')
plt.xlabel('Count')
plt.show()
```

## 5. Correlations & Scatter Analysis

```python
# Correlation between numerical columns
corr = df_biz[["stars", "review_count", "latitude", "longitude", "checkin_count", "tip_count"]].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## 6. Time Series Analysis

Visualizing monthly user check-ins and tip trends over time.

```python
# Parse year_month and group
df_ts['year_month'] = pd.to_datetime(df_ts['year_month'])
df_monthly = df_ts.groupby('year_month')[['checkins', 'tips']].sum().reset_index()

fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(df_monthly['year_month'], df_monthly['checkins'], color='tab:blue', label='Check-ins')
ax1.set_xlabel('Date')
ax1.set_ylabel('Check-ins', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.plot(df_monthly['year_month'], df_monthly['tips'], color='tab:orange', linestyle='--', label='Tips')
ax2.set_ylabel('Tips', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

plt.title('Monthly Trends of Check-ins and Tips Over Time')
fig.tight_layout()
plt.show()
```

## Conclusion & Key Insights

1. **Rating Distribution:** The distribution of star ratings is skewed towards positive ratings, peaking at 4.0 stars.
2. **Geographical Concentration:** Pennsylvania (PA) and Florida (FL) represent the states with the highest density of businesses in the dataset. Philadelphia is the top city.
3. **Dominant Industries:** Restaurants, Food, and Shopping are the primary drivers of business entries and user engagement.
4. **Correlations:** There is a moderate positive correlation between check-in counts and tip counts, suggesting that businesses with high visit volumes also receive high user feedback volume.
